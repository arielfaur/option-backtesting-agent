# Import data from csv

```python
import os
import pandas as pd


# Read each file into a DataFrame and store them in a list
dfs = [pd.read_csv(f"./data/spx_eod_2023{i:02}.txt", parse_dates=True, low_memory=False) for i in range(1,13)]

# Concatenate the DataFrames
df_concat = pd.concat(dfs)

print(df_concat.head())  # Print the first few rows of the concatenated DataFrame
```

```python
df_concat
```

```python
df_concat.columns = df_concat.columns.map(lambda x: str(x).strip().strip('[]').lower())
```

```python
df_concat.c_iv = pd.to_numeric(df_concat['c_iv'], errors='coerce') 
df_concat.p_iv = pd.to_numeric(df_concat['p_iv'], errors='coerce') 
df_concat.c_volume = pd.to_numeric(df_concat['c_volume'], errors='coerce') 
df_concat.p_volume = pd.to_numeric(df_concat['p_volume'], errors='coerce') 
df_concat.quote_readtime = pd.to_datetime(df_concat['quote_readtime'], errors='coerce') 
df_concat.quote_date = pd.to_datetime(df_concat['quote_date'], errors='coerce') 
df_concat.expire_date = pd.to_datetime(df_concat['expire_date'], errors='coerce')
df_concat.c_delta = pd.to_numeric(df_concat['c_delta'], errors='coerce')
df_concat.p_delta = pd.to_numeric(df_concat['p_delta'], errors='coerce')
df_concat.c_gamma = pd.to_numeric(df_concat['c_gamma'], errors='coerce')
df_concat.p_gamma = pd.to_numeric(df_concat['p_gamma'], errors='coerce')
df_concat.c_theta = pd.to_numeric(df_concat['c_theta'], errors='coerce')
df_concat.p_theta = pd.to_numeric(df_concat['p_theta'], errors='coerce')
df_concat.c_vega = pd.to_numeric(df_concat['c_vega'], errors='coerce')
df_concat.p_vega = pd.to_numeric(df_concat['p_vega'], errors='coerce')
df_concat.c_rho = pd.to_numeric(df_concat['c_rho'], errors='coerce')
df_concat.p_rho = pd.to_numeric(df_concat['p_rho'], errors='coerce')
df_concat.c_bid = pd.to_numeric(df_concat['c_bid'], errors='coerce')
df_concat.p_bid = pd.to_numeric(df_concat['p_bid'], errors='coerce')
df_concat.c_ask = pd.to_numeric(df_concat['c_ask'], errors='coerce')
df_concat.p_ask = pd.to_numeric(df_concat['p_ask'], errors='coerce')
df_concat.c_last = pd.to_numeric(df_concat['c_last'], errors='coerce')
df_concat.p_last = pd.to_numeric(df_concat['p_last'], errors='coerce')
```

```python
df_concat.info()
```

```python
df_concat.drop(columns=['quote_unixtime', 'quote_readtime', 'quote_time_hours', 'expire_unix'], inplace = True)
```

```python
df = df_concat.set_index('quote_date')
```

```python
df.to_csv('data/spx_eod_2023.csv', index=True)
```
